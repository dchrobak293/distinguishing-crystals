## Figure 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = './kumagai/enthalpy/'

dia  = np.loadtxt(path+'dia.dat')
btin = np.loadtxt(path+'btin.dat')
bct5 = np.loadtxt(path+'bct5.dat')
bc8  = np.loadtxt(path+'bc8.dat')

cm = 1/2.54  # centimeters in inches
fig, axs = plt.subplots(2,1,figsize=(15*cm, 20*cm),layout="constrained")

axs[0].plot(dia[:,0],dia[:,1], color='r', linewidth=1)
axs[0].plot(bct5[:,0],bct5[:,1], color='g', linewidth=1)
axs[0].plot(btin[:,0],btin[:,1], color='b', linewidth=1)
axs[1].plot(btin[:,0],btin[:,1], color='b', linewidth=1)
axs[1].plot(bc8[:,0],bc8[:,1], color='m', linewidth=1)

axs[0].set_xlim(9,18)
axs[0].set_ylim(-3.7,-2.7)
axs[1].set_xlim(4,10)
axs[1].set_ylim(-4,-3.3)

lb=['a','b']

for i in range(2):
    axs[i].xaxis.set_tick_params(labelsize=12)
    axs[i].yaxis.set_tick_params(labelsize=12)
    axs[i].text(-0.12, 1.05, lb[i], transform=axs[i].transAxes, size=14, weight='bold')
    axs[i].set_ylabel('Enthalpy [eV]', fontsize=13);
    axs[i].set_xlabel('Pressure [GPa]', fontsize=13)

plt.savefig('Fig_S1_0.pdf',bbox_inches='tight')
```
